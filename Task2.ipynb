# Calculate the total weight of each product

```python
sum_kg_of_each_product_id = {}
for productID in new_df['ProductID'].unique().tolist():
    result = new_df[new_df['ProductID']==productID]["DispatchQuantityInKg"].sum()
    sum_kg_of_each_product_id[productID] = result
print(sum_kg_of_each_product_id)
```

```python
for key,value in sum_kg_of_each_product_id.items():
    print(f"{key}: {value} kg")
```

```python
def map_value(value):
    dct = {
        value == "Carton": "CTN",
        value == "Plastic Bottle": "PBOT",
        value == "Glass Bottle": "GBOT"
    }
    for key, value in dct.items():
        if key:
            return value
```
